# 2 Plotting a Function
Goal: Write a **function** that will plot $f(x) = x^2$ and $g(x) = \sqrt{x}$ on the same plot and return the $f(x)$ and $g(x)$ values as separate lists. Your plot needs to have the following:
- An appropriate figure size (`figsize=(10,8)` is recommended)
- Labeled x and y axis
- A title
- Different colors and linestyles for each function
- A legend
- Your domain should be $x \in [0, 10]$. *Hint: Use `np.linspace` to set a range of values for your functions.*

Your function should return:
- A list of $f(x)$ values
- A list of $g(x)$ values.

```python
def plot_x_squared_and_root_x():
    # Plots x squared and x to the 1/2 on the same graph with different color lines.
    plt.figure(figsize=(10,8))
    x_data = np.linspace(0,10,1000)
    y_1 = x_data**2
    y_2 = x_data**(1/2)
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("graph of f(x) = x^2 and g(x) = x^1/2")
    plt.plot(x_data, y_1, label = "f(x) = x^2", linestyle="-", color = "red")
    plt.plot(x_data, y_2, label = "g(x) = x^1/2", linestyle="--", color = "green")
    plt.legend()
    plt.show()
    return y_1, y_2

plot_x_squared_and_root_x()
```

```python
def plot_a_3D_sphere(radius = 1):
    # Makes a sphere by creating a list of x and y values ranging from the negative to positive radius and then using the sphere formula to get the top half of a sphere(only plots positive z). 
    x = np.linspace(-radius, radius, 1000)
    y = np.linspace(-radius, radius, 1000)
    z = (radius**2 - x**2 - y**2)**0.5
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.plot(x,y,z, label = "z = (r^2-x^2-y^2)^0.5)")
    ax.set_xlabel("x-axis")
    ax.set_ylabel("y-axis")
    ax.set_zlabel("z-axis")
    plt.title("sphere cross section")
    plt.legend()
    plt.show()
    return

plot_a_3D_sphere()
```
